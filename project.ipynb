#### Normalize the variable of prices for each population and plot the 4 histograms in the same plot (★★★) 

```python
## Get the Prices for each Population
fuenlabrada_prices = ds_sbp_subset[ds_sbp_subset['level5']=='Fuenlabrada']['price']

leganes_prices = ds_sbp_subset[ds_sbp_subset['level5']=='Leganés']['price']

getafe_prices = ds_sbp_subset[ds_sbp_subset['level5']=='Getafe']['price']

alcorcon_prices = ds_sbp_subset[ds_sbp_subset['level5']=='Alcorcón']['price']

##Normalizing using Z-Score Normalization to normalize with a mean of 0 and a Standar Deviation of 1
fuenlabrada_prices_mean = fuenlabrada_prices.mean()
fuenlabrada_prices_std = fuenlabrada_prices.std()
fuenlabrada_prices_normalized = (fuenlabrada_prices - fuenlabrada_prices_mean)/fuenlabrada_prices_std

leganes_prices_mean = leganes_prices.mean()
leganes_prices_std = leganes_prices.std()
leganes_prices_normalized = (leganes_prices - leganes_prices_mean)/leganes_prices_std

getafe_prices_mean = getafe_prices.mean()
getafe_prices_std = getafe_prices.std()
getafe_prices_normalized = (getafe_prices - getafe_prices_mean)/getafe_prices_std

alcorcon_prices_mean = alcorcon_prices.mean()
alcorcon_prices_std = alcorcon_prices.std()
alcorcon_prices_normalized = (alcorcon_prices - alcorcon_prices_mean)/alcorcon_prices_std


#Plot the Histograms of each Normalized price variable in the same plot.
fig, ax = plt.subplots(2, 2, sharey = True)

ax[0, 0].hist(fuenlabrada_prices_normalized)
ax[0, 1].hist(leganes_prices_normalized)
ax[1, 0].hist(getafe_prices_normalized)
ax[1, 1].hist(alcorcon_prices_normalized)
plt.show()
```

#### What can you say about the price per square meter between the towns of Getafe and Alcorcón?

```python
# To get the relation between the price per sqare meter of Getafe and Alcorcon, the mean of the price per square meter of each population will be
# calculated by separated. Then the means will be compared.

##Get the Price and the Surface of Getafe: 
#Get a set with the price per square meter for each house of Getafe (with non NAN surface value).
price_psm_getafe = (ds_sbp_subset[ds_sbp_subset['level5']=='Getafe']['price'])/(ds_sbp_subset[ds_sbp_subset['level5']=='Getafe']['surface'])

#Calculate the mean of the price per square meter of the houses of Getafe
price_psm_getafe_mean = price_psm_getafe.mean()

print("The mean of the price per square meter of Getafe is :" + str(price_psm_getafe_mean))

#### FIRST RESULT: The mean of the price per square meter of getafe is 2066.314949251463 Euros

#### Do the same process for Alcorcon
#Get a set with the price per square meter for each house of Alcorcon (with non NAN surface value).
price_psm_alcorcon = (ds_sbp_subset[ds_sbp_subset['level5']=='Alcorcón']['price'])/(ds_sbp_subset[ds_sbp_subset['level5']=='Alcorcón']['surface'])

#Calculate the mean of the price per square meter of the houses of Alcorcon
price_psm_alcorcon_mean = price_psm_alcorcon.mean()

print("The mean of the price per square meter of Alcorcon is :" + str(price_psm_alcorcon_mean))

#### SECOND RESULT: The mean of the price per square meter of getafe is 2239.302480199618 Euros

##Calculate the percentaje of diference between the price per square meter of Getafe vs Alcorcon

percentage = np.abs((price_psm_alcorcon_mean - price_psm_getafe_mean))/ price_psm_getafe_mean * 100

################ FINAL RESULT: ################
print("The price per square meter of Alcorcon is " + str(percentage) + "% more expensive than the price per square meter of Getafe")

###The price per square meter of Alcorcon is 8.37% more expensive than the price per square meter of Getafe
```

#### Reload the data, is the average of "Valdemorillo" and "Galapagar" prices the same?

#### What happens if we do the same analysis for price/m2 (★★☆) 

```python
##Get the price per square meter of Valdemorillo
price_per_sqm_valdemorillo = (ds[ds['level5']=='Valdemorillo']['price'])/(ds[ds['level5']=='Valdemorillo']['surface'])

price_per_sqm_valdemorillo_mean = price_per_sqm_valdemorillo.mean()

print("The average of the price per square meter of Valdemorillo is: " + str(price_per_sqm_valdemorillo_mean))

##Get the price per square meter of Galapagar
price_per_sqm_galapagar = (ds[ds['level5']=='Galapagar']['price'])/(ds[ds['level5']=='Galapagar']['surface'])

price_per_sqm_galapagar_mean = price_per_sqm_galapagar.mean()

print("The average of the price per square meter of Galapagar is: " + str(price_per_sqm_galapagar_mean))


###RESULT
percentage = np.abs((price_per_sqm_galapagar_mean - price_per_sqm_valdemorillo_mean))/ price_per_sqm_valdemorillo_mean * 100

################ FINAL RESULT: ################
print("The price per square meter of Valdemorillo is " + str(percentage) + "% more expensive than the price per square meter of Galapagar")

###The price per square meter of Valdemorillo is 21.88% more expensive than the price per square meter of Galapagar.
### The average of the price per square meter of Valdemorillo and Galapagar are NOT the same.
```
